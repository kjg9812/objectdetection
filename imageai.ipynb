```python
#static images
from imageai.Detection import ObjectDetection
detector = ObjectDetection()
model_path = "./model/yolov3.pt"
input_path = "./input/people.jpg"
output_path = "./output/newimage.jpg"
detector.setModelTypeAsYOLOv3()
detector.setModelPath(model_path) 
detector.loadModel()  
detection = detector.detectObjectsFromImage(input_image=input_path, output_image_path=output_path, minimum_percentage_probability=30)
for eachObject in detection:
    print(eachObject["name"] , " : ", eachObject["percentage_probability"], " : ", eachObject["box_points"] )
    print("--------------------------------")
```

```python
#videos
from imageai.Detection import VideoObjectDetection
detector = VideoObjectDetection()
model_path = "./model/yolov3.pt"
input_path = "./input/cars.mp4"
output_path = "./output/newvideo"
detector.setModelTypeAsYOLOv3()
detector.setModelPath(model_path)
detector.loadModel()


def forFrame(frame_number, output_array, output_count):
    print("FOR FRAME " , frame_number)
    print("Output for each object : ", output_array)
    print("Output count for unique objects : ", output_count)
    print("------------END OF A FRAME --------------")

video_path = detector.detectObjectsFromVideo(input_file_path=input_path,
                            output_file_path=output_path
                            , frames_per_second=20, log_progress=False,per_frame_function=forFrame,minimum_percentage_probability=30)
print(video_path)
```
